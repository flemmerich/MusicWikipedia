# Correlation betweeen Hofstede dimensions and music genres

```python
import numpy as np
import pandas as  pd
```

#### Preprocesssing

Load genre data

```python
csv = pd.read_csv("C:/data/MusicWikipedia/LFM-1b_country_genre_distributions_allmusic.txt", sep="\t", names=["country", "genre", "share"], header=None)
```

Convert data into table

```python
genres = list(csv.genre.unique())
countries = list(csv.country.unique())

dataTable = np.zeros((len(countries), len(genres)))
for index, row in csv.iterrows():
    
    country = row["country"]
    genre = row ["genre"]
    dataTable[countries.index(country), genres.index(genre)] = row["share"]
df = pd.DataFrame(data=dataTable, columns = genres)
df["ctr"] = countries

df
```

Load & preprocess Hofstede data

```python
#load
hofstede = pd.read_csv("C:/data/MusicWikipedia/hofstede.txt", sep="\t")

#set index column
hofstede.set_index(['ctr'])

#remove columns that we do not need
del hofstede['no']
del hofstede['country']

hofstede
```

Merge both data frames and show result

```python
mergeData = hofstede.merge(df, on="ctr")
mergeData
```

Compute pairwise correlation,

Filter result matrix such that it only contains rows/columns that correlate a Hofstede dimension with a genre

```python
correlation_matrix = mergeData.corr(method='spearman') [1:7]
correlation_matrix = correlation_matrix.transpose()[7:]
correlation_matrix
```

Visualize Correlation Matrix

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.colors as mcolors
%matplotlib inline
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)

#color palette
cmap = mcolors.LinearSegmentedColormap.from_list(name='red_white_blue', 
                                                 colors =[(0., 0., 1.), 
                                                          (1., 1., 1.), 
                                                          (1., 0., 0.)],
                                                 N=100,
                                                 )
cax = ax.imshow(correlation_matrix, interpolation='nearest', cmap=cmap, vmin=-0.6, vmax=0.6)
fig.colorbar(cax)

ax.set_xticklabels(["", "PD", "IND", "MAS", "UNC", "LTO", "IDU"])
ax.set_yticklabels([""] + genres)
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))

plt.show()
```

```python
def scatter(attX, attY, labelX, labelY):
    labels = mergeData.ctr

    plt.plot(attX, attY, "o")

    for label, x, y in zip(labels, attX, attY):
        plt.annotate(
            label, 
            xy = (x, y), xytext = (5, 5),
            textcoords = 'offset points', ha = 'right', va = 'bottom'
            #arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0')
        )

    plt.xlabel(labelX)
    plt.ylabel(labelY)
    plt.show()
```

Visualize some example correlations

```python
scatter(mergeData.individualism, mergeData.rap, "Individualism", "Rap")
```

```python
scatter(mergeData.long_term_orientation, mergeData.jazz, "Long Term Orientation", "Jazz")
```

### Quality of Government data
http://qog.pol.gu.se/data/datadownloads/qogbasicdata


```python
#load
qog = pd.read_csv("C:/data/MusicWikipedia/qog.csv", sep=",", quotechar='"')
scoecoData = mergeData.merge(qog, on="ctr", how="left")
```

```python
soceco_features = ["une_gdpc", "rsf_pfi", "fe_cultdiv", "undp_hdi", "une_rp"]
soceco_features_labels = ["GDP", "Press Freedom Index", "Cultural_Diversity", "Human Development Index", "Rural population"]
corr_data = scoecoData[ soceco_features + genres]
```

```python
correlation_soceco = corr_data.corr(method='spearman')  [0:5]
correlation_soceco = correlation_soceco.transpose()[5:]
correlation_soceco
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)

#color palette
cmap = mcolors.LinearSegmentedColormap.from_list(name='red_white_blue', 
                                                 colors =[(0., 0., 1.), 
                                                          (1., 1., 1.), 
                                                          (1., 0., 0.)],
                                                 N=100,
                                                 )
cax = ax.imshow(correlation_soceco, interpolation='nearest', cmap=cmap, vmin=-0.6, vmax=0.6)
fig.colorbar(cax)

ax.set_xticklabels([""] + soceco_features_labels, rotation=90)
ax.set_yticklabels([""] + genres)
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))

```
